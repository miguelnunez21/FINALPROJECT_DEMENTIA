```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Crear un OneHotEncoder para la columna 'dementia_all'
onehot_encoder = OneHotEncoder(drop='first', sparse_output=False)

# Aplicar One-Hot Encoding a la columna 'dementia_all'
dementia_encoded = onehot_encoder.fit_transform(dem_standardized[['dementia_all']])

# Convertir a DataFrame para facilitar la visualización
dementia_encoded_df = pd.DataFrame(dementia_encoded, columns=onehot_encoder.get_feature_names_out(['dementia_all']))

# Crear un PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)

# Aplicar expansión polinómica
dementia_poly = poly.fit_transform(dementia_encoded_df)

# Convertir a DataFrame para facilitar la visualización
dementia_poly_df = pd.DataFrame(dementia_poly, columns=poly.get_feature_names_out(onehot_encoder.get_feature_names_out(['dementia_all'])))
# Tomar solo una de las nuevas columnas como la columna 'dementia_all' 
dementia_poly_df_single = dementia_poly_df.iloc[:, 0] # Usamos la primera columna

# Añadir la nueva columna 'dementia_all' al DataFrame original 
dem_standardized['dementia_all'] = dementia_poly_df_single

print (dem_standardized)





dem_standardized_encoded = pd.get_dummies(dem_standardized, drop_first=True)
print (dem_standardized_encoded)
```

```python
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input 
from sklearn.model_selection import KFold

X = dem_standardized_encoded.drop('dementia_all', axis=1)
y = dem_standardized_encoded['dementia_all']

input_shape = X.shape[1]
# Define the number of folds for cross-validation
n_splits = 10
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize an empty list to store the accuracy scores for each fold
cv_scores = []

# Iterate through the folds
for train_index, val_index in kf.split(X):
    # Split the data into training and validation sets for the current fold
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    
    def create_model():
  # Create a sequential model
      model = Sequential([
        Input(shape =(input_shape,)),
        Dense(64, activation='relu'),
        #Dense(100, activation='relu'),
        Dense(1, activation='sigmoid')
      ])
      model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
      return model
    

    # Create a new model for each fold
    model = create_model()

    # Train the model on the training data
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
                                   
    model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0, validation_data=(X_val, y_val), callbacks=[early_stopping]) # Set verbose to 0 to suppress output

    # Evaluate the model on the validation data
    _, accuracy = model.evaluate(X_val, y_val, verbose=0)
    cv_scores.append(accuracy)

# Print the accuracy scores for each fold
for fold, score in enumerate(cv_scores):
    print(f"Fold {fold+1}: Accuracy = {score*100:.2f}%")

# Calculate and print the average accuracy across all folds
mean_accuracy = np.mean(cv_scores)
std_deviation = np.std(cv_scores)
print(f"Average Accuracy across all folds: {mean_accuracy*100:.2f}% ± {std_deviation*100:.2f}%")
```

```python
test_loss, test_acc = model.evaluate(X_val, (y_val))
print(f"Test accuracy: {test_acc}")
```
