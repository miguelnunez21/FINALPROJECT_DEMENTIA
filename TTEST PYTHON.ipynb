```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt      

#MODEL 1= KNN, MODEL 2 = NN, MODEL 3= DLNN
model_means = [90.13, 91.18, 90.54]
model_sds = [2.19, 2.02, 1.64]
model_names = ["k-NN", "NN", "DLNN"]
samples_model_1 = np.random.normal(90.13, 2.19, 10)
samples_model_2 = np.random.normal(91.18, 2.02, 10)
samples_model_3 = np.random.normal(90.54, 1.64, 10)

# Inicializar la matriz vacía
comparison_matrix = pd.DataFrame(index=model_names, columns=model_names)

# Llenar la matriz
for i in range(len(model_names)):
    for j in range(len(model_names)):
        if i == j:
            # Diagonal principal: Mostrar media ± SD
            comparison_matrix.iloc[i, j] = f"{model_means[i]:.3f} ± {model_sds[i]:.3f}"
        elif i < j:
            # Celdas superiores: Calcular t-test pareado y mostrar p-value
            t_stat, p_value = stats.ttest_rel(eval(f"samples_model_{i+1}"), eval(f"samples_model_{j+1}"))
            comparison_matrix.iloc[i, j] = f"{p_value:.5f}"
        else:
            # Celdas inferiores: Vacías
            comparison_matrix.iloc[i, j] = ""

comparison_matrix = pd.DataFrame(index=model_names, columns=model_names)

# Fill the matrix
for i in range(len(model_names)):
    for j in range(len(model_names)):
        if i == j:
            # Diagonal principal: Mostrar media ± SD
            comparison_matrix.iloc[i, j] = f"{model_means[i]:.3f} ± {model_sds[i]:.3f}"
        elif i < j:
            # Celdas superiores: Calcular t-test pareado y mostrar p-value
            t_stat, p_value = stats.ttest_rel(eval(f"samples_model_{i+1}"), eval(f"samples_model_{j+1}"))
            comparison_matrix.iloc[i, j] = f"{p_value:.5f}"
        else:
            # Celdas inferiores: Vacías
            comparison_matrix.iloc[i, j] = ""

comparison_matrix


# Mostrar la matriz generada
#print(comparison_matrix.to_string())
```
